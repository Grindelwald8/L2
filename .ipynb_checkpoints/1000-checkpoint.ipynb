```python
def pixels(name):
    img = Image.open("{}".format(name))
    obj = img.load()
    mas2 = []
    k = 0
    # print(obj[255,0])
    mas2.append(1)
    for i in range(0, 100):
        # print("i = "+str(i))
        for j in range(0, 100):
            # print("j="+str(j))
            mas2.append(obj[i, j])
            j = j + 1
            k = k + 1
        i = i + 1
    # print(k)
    #print(len(mas2))
    return mas2
```

```python
training_data_list = []
import pathlib
# определение пути
#currentDirectory = pathlib.Path('Frozen')
#for currentFile in currentDirectory.iterdir():  
    #print(currentFile)
currentFile = "C:/Users/User/Desktop/MISIS_1/Neuro/L2/Frozen"
currentFile = os.path.normpath(currentFile)
print(currentFile)
for i in range(0, len(files)):
    #print(i)
    #print(files[i])
    img = Image.open(os.path.join(currentFile,files[i]))
    #print(os.path.join(currentFile,files[i]))
    img = img.resize((100, 100), Image.ANTIALIAS)
    img = img.convert('L')
    img.save(os.path.join(currentFile,files[i]))
    mas12 = pixels(os.path.join(currentFile,files[i]))
    training_data_list.append(mas12)
    mas12 = []
```
